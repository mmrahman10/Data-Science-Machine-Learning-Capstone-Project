## Abstract
Which type of complaint should the Department of Housing Preservation and Development of New York City focus on first?
Should the Department of Housing Preservation and Development of New York City focus on any particular set of boroughs, ZIP codes, or street (where the complaints are severe) for the specific type of complaints identified?
Does the Complaint Type identified have an obvious relationship with any particular characteristic or characteristics of the houses or buildings?
Can a predictive model be built for a future prediction of the possibility of complaints of the type that have identified?

For this project, I used logistic regression, K nearest neighbors classifer, Decision tree, SVM, Naive Bayes classifier and a gradient boosting classifier. Overall, the model that most accurately predicted was the gradient boosting classifer with an accuracy of 73.19%. The gradient boosting classifer was able to paint a picture of the importances of the different building features that impact complaints type most notably BldgDepth','BuiltFAR', 'LotDepth', 'NumFloors', 'ResArea', and building age('BldAge').

<h2 id="indicator">Indicator variable (or dummy variable)</h2>
<b>What is an indicator variable?</b>
<p>
    An indicator variable (or dummy variable) is a numerical variable used to label categories. They are called 'dummies' because the numbers themselves don't have inherent meaning. 
</p>

<b>Why we use indicator variables?</b>
<p>
    So we can use categorical variables for regression analysis in the later modules.
</p>
<b>Example</b>
<p>
    In this model I used two top "Complaint Type" , "HEAT/HOTWATER" and "HEATER". Regression doesn't understand words, only numbers. To use this attribute in regression analysis, we convert "Complaint Type" into indicator variables. This will work only for Binary Classification.
</p>

<p>
    We will use the panda's method 'get_dummies' to assign numerical values to different categories of Complaint Type.
    <b>BUT BETTER TO USE sklearn's LabelEncoder. IN THIS CASE WE DONT NEED TO CONCATE THE COLUMN WITH DATAFRAME. BUT FOR SOME REASON IT START WITH 1,2,3,4 AND END WITH 0 IN MY CASE!!</b>
</p>
